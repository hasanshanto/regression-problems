# **Support Vector Regression Template**




## **Importing Necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Importing the Dataset**

```python
dataset = pd.read_csv('Data.csv')

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
y = y.reshape(len(y), 1)
```

## **Splitting the dataset into Train and Test Data**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state= 42)
```

## **Feature Scalling the Data**

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
sc_y = StandardScaler()

X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

## **Training the model using Train Dataset**

```python
from sklearn.svm import SVR

regressor = SVR(kernel= 'rbf')
regressor.fit(X_train, y_train)
```

## **Predicting and Comparing the model with Test Data**

```python
y_pred1 = regressor.predict(sc_X.transform(X_test))
```

```python
y_pred = sc_y.inverse_transform(y_pred1.reshape(len(y_pred1), 1))
```

```python
np.set_printoptions(precision= 2)

np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test), 1)), axis= 1)
```

## **Evaluating the Models Performance**

```python
from sklearn.metrics import r2_score

r2_score(y_pred, y_test)
```
